# Visualizacion de datos con Python: Matplotlib, Seaborn y Plotly


```python
# importar librerias
import pandas as pd
```

```python
#cargar archivo dataframe
df = pd.read_csv('inmigrantes_canada.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# Cuantos registros tiene
df.shape
```

```python
#todas tienen 195 registros non-null significa que contiene un dato valido
df.info()
```

```python
#vamos utilizar metodo set_index y indicamos indice pais, y hacemos modificacion inplace sobre el mmismo dataframe, pais inicia como indice
df.set_index('Pais', inplace=True)
df
```

```python
#generar un nuevo un dataframe lista de años y cantidad de emigrantes de colombia, lista de strings
anos = list(map(str,range(1980,2014)))
anos
```

```python
#aplicar este filtro a colombia
colombia =df.loc['Colombia', anos]  #ano, cantidad de emigrantes
colombia
```

```python
#Crear diccionario
col_dict ={'Año':colombia.index.tolist(), 'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
#Creamos un nuevo DataFrame y guardamos nuevos datos
datos_col =pd.DataFrame(col_dict)
datos_col.tail()
```

## Crear graficos con Matplotlib

```python
#Crear nuevs graficas importamos librerias
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
```

```python
#Ajustes
plt.figure(figsize=(8,4))# 8 pulgadas de ancho por 4 pulgadas de alto
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])# seleccionamos los tipos de datos
plt.xticks(['1980','1985', '1990','1995', '2000','2005', '2010']) #tomamos datos de los años espaciado entre 5 y 5
plt.title('Inmigracion de Colombianos a Canada')# Titulo de la grafica
plt.xlabel('Año')# Etiqueando los label x y y
plt.ylabel('Numero de Inmigrantes')
plt.show()# presentamos la grafica
```

```python
#Como almacenar todas estas variables fig es la variable
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de Colombianos a Canada 1980 -2013')
ax.set_xlabel('Año')
ax.set_ylabel('Numero de Inmigrantes')
plt.show()
```

```python
# obtener mas datos con otro grafico por ejemplo grafico de caja fig es el lienzo
fig, axs = plt.subplots(1,2,figsize=(10,3))#una fila dos graficas
axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])# grafica izquierda = 0
axs[0].set_title('Inmigrantes de Colombianos hacia Canada \n1980 -2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de <inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmiracion de Colombianos \nhacia Canada 1980 -2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de Inmigrantes')
plt.show()
```

```python
#Adicionar cuadricula
# obtener mas datos con otro grafico por ejemplo grafico de caja fig es el lienzo
fig, axs = plt.subplots(1,2,figsize=(10,3))#una fila dos graficas
axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])# grafica izquierda = 0
axs[0].set_title('Inmigrantes de Colombianos hacia Canada \n1980 -2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de <inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmiracion de Colombianos \nhacia Canada 1980 -2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de Inmigrantes')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Numero de Inmigrantes')
plt.show()
```

```python
#poner escalas iguales en todas las graicas
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigracion de los cuatro mayores paises sudamericanos\nhacia Canada desde 1980 - 2013')

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin=0
ymax=7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Numero de Inmigrantes')
  ax.grid()


plt.show()
```

```python
#Como almacenar todas estas variables fig es la variable, cambio tamaño de la fuente con fontsize
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'],lw=2,marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de Colombianos a Canada\n1980 -2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Numero de Inmigrantes')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid()
plt.show()
```

```python
#uso de colores
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'],lw=2,color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de Colombianos a Canada\n1980 -2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Numero de Inmigrantes')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle= '--')
plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
```

```python
sudamerica.head(10)
```

```python
#Graficar los totales inmigragantes de estos paises
colores =['royalblue','orange','forestgreen', 'orchid','purple','brown','slateblue','gray','olive','navy','teal','tomato']
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo 1980 hasta 2013', loc= 'left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
#organizar el grafico
fig, ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo 1980 hasta 2013', loc= 'left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)
#Graficar los totales inmigragantes de estos paises
colores =['royalblue','orange','forestgreen', 'orchid','purple','brown','slateblue','gray','olive','navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo 1980 hasta 2013', loc= 'left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores =[]
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
      colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamerica: Colombia es el segundo pais con mas migrantes hacia Canada\ndurante el periodo 1980 hasta 2013', loc= 'left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)


plt.show()
```

```python
#Creando anotaciones
colores =[]
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
      colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))
#Eliminar el frame o cuadro
ax.set_frame_on(False)
ax.xaxis.set_visible(False)

ax.tick_params(axis='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamerica: Colombia es el segundo pais con mas migrantes hacia Canada\ndurante el periodo 1980 hasta 2013', loc= 'left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Numero de Inmigrantes', fontsize=14)
#ax.xaxis.set_tick_params(labelsize=12)
#ax.yaxis.set_tick_params(labelsize=10)
#anotaciones
for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j), color='black', fontsize=8.5, ha='left',va='center')

#almacenar
fig.savefig('inmigrantes_suramericanos_canada.png', transparent=False, dpi=300,bbox_inches='tight')

plt.show()
```

## Crear graficos con Seaborn

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
sns.barplot(data=top_10, x='Total', y=top_10.index,orient='h')
ax.set(title='Top 10 paises con mayor migracion hacia Canada\n1980 -2013',xlabel='Numero de Inmigrantes',ylabel='Año')

plt.show()
```

```python
#combinar seaborne con matplotlib
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index,orient='h')
ax.set_title('Top 10 paises con mayor migracion hacia Canada\n1980 -2013', loc='left', fontsize=16)
ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
# cambiar colores a las barras
def generar_grafico(palette):

  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, x='Total', y= top_10.index, hue=top_10.index,orient='h',palette=palette, legend=False  )
  ax.set_title('Top 10 paises con mayor migracion hacia Canada\n1980 -2013', loc='left', fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')

  plt.show()
```

## Herramientas para graficos mas dinamicos Plotly

```python
import plotly.express as px
```

```python
#creamos una figura con fig
fig = px.line(datos_col, x='Año', y='Inmigrantes')
fig.update_layout(width=1000, height=500,xaxis={'tickangle': -45})
fig.show()
```

```python
#Añadir rotulos y modificar fuentes
fig = px.line(datos_col, x='Año', y='Inmigrantes', title='Inmigracion de Colombianos hacia Canada entre 1980 y 2013')
fig.update_layout(width=800, height=500,xaxis={'tickangle': -45},xaxis_title='Año', yaxis_title='Numerode Inmigrantes', font_family='Arial',font_size=14, font_color='grey')

fig.show()
```

```python
# Cambiar color y ancho de la linea
fig = px.line(datos_col, x='Año', y='Inmigrantes', title='Inmigracion de Colombianos hacia Canada entre 1980 y 2013')
fig.update_traces(line_color='green', line_width=3)
fig.update_layout(width=800, height=500,xaxis={'tickangle': -45},xaxis_title='Año', yaxis_title='Numerode Inmigrantes', font_family='Arial',font_size=14, font_color='grey')

fig.show()
```
