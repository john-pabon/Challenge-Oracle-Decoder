## Transformacion de datos de tipo texto

```python
#primer paso de tokenizacion llevarlo a miniscula
#Palabras que le pueden aportar, impactar un valor al precio y se les puede dar un valor, pueden ser parametros de estudio
datos['descripcion_local']= datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
#Eliminar texto que no nos transmite tanta informacion
datos['descripcion_local'][3169]
```

```python
#Eliminar texto que no nos transmite tanta informacion, uso de Regex
#datos['descripcion_local'].str.replace('palabra a buscar',' ', regex=True)
#que significa esto = [^a-zA-Z0-9\-\'] esto significa que no sustituya en letras de a-z, A-Z, numeros de 0-9
# ^ => negacion remplazar todo aquello menos lo que esta aqui a-zA-Z0-9\-\ remplace todo lo otro
# \-\' => no substituya los guiones ni las comillas simples, todo el resto remplace por espacioo en blanco
datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
#guardar los datos modificados
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
# otro regex para palabras no compuestas, remplazo de guiones
# esto significa (?<!\w)-(?!\w)=> remplaze los guiones pero que no tenga un caracter a la izq y busca que no tenga
#caracter a la derecha, esto da True lo sustituye por nada ej one - shot
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
```

```python
datos['descripcion_local'].head()
```

```python
#Tokenizacion de strings, creamos una lista de palabras
datos['descripcion_local'].str.split()
```

```python
# guardar los datos
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
# Regex para otra columnna
# \{|}|\"'=> busque la llave de apertura, barra vertical = utiliza simbolo separador y busque otra cosa
#busque la llave de cierre, barra vertical busque las comillas dobles, se cierre y se remplaza por nada
datos['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
# guardar en una lista
datos['comodidades']=datos['comodidades'].str.split(',')
```

```python
#guardar datos
datos['comodidades']=datos['comodidades'].str.replace('\{|}|\"','',regex=True)
```
